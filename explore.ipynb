# Data Exploration - WIPO Innovation Capabilities 2026

This notebook provides a quick overview and verification of the datasets provided in the `data/` folder. We will check the structure of places (units), innovation fields, and the actual innovation outputs.

```python
import pandas as pd
import numpy as np

def print_info(name, df):
    print(f"--- {name} ---")
    print(f"Shape: {df.shape}")
    print(f"Columns: {list(df.columns)}")
    display(df.head())
    print("\n")
```

## 1. Load Reference Data
We load the information about countries (units) and innovation fields.

```python
units = pd.read_parquet('data/units.parquet')
fields = pd.read_parquet('data/fields.parquet')

print_info("Units (Places)", units)
print_info("Fields (Innovation Categories)", fields)
```

## 2. Load Innovation Outputs
The `outputs.parquet` file contains the volume of innovation (patents, papers, etc.) by period, place, and field.

```python
outputs = pd.read_parquet('data/outputs.parquet')
print_info("Innovation Outputs", outputs)
```

## 3. Data Verification
Check for missing values and unique counts to ensure data integrity.

```python
print("Missing values in Outputs:")
print(outputs.isnull().sum())

print("\nUnique counts:")
print(f"Unique Periods: {outputs['Period'].nunique()}")
print(f"Unique Units: {outputs['Unit'].nunique()}")
print(f"Unique Fields: {outputs['Field ID'].nunique()}")
print(f"Unique Dimensions: {outputs['Dimension'].unique()}")

print("\nCounts by Dimension:")
print(outputs['Dimension'].value_counts())
```
